```python

import numpy as np
from sklearn.metrics import average_precision_score
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.io
import pandas as pd
```

```python
def txtData_to_array(path):
    with open(path) as file:
        lines = file.readlines()
        lines = [float(line.rstrip()) for line in lines]
        arr = np.array(lines)
    return arr
```

```python
scores_online_wine = txtData_to_array('online_data/online_wine.txt')
scores_online_lympho = txtData_to_array('online_data/online_lympho.txt')
scores_online_glass = txtData_to_array('online_data/online_glass.txt')
scores_online_ionosphere = txtData_to_array('online_data/online_ionosphere.txt')
scores_online_pima = txtData_to_array('online_data/online_pima.txt')
scores_online_mnist = txtData_to_array('online_data/online_mnist.txt')

scores_offline_wine = txtData_to_array('offline_data/offline_wine.txt')
scores_offline_lympho = txtData_to_array('offline_data/offline_lympho.txt')
scores_offline_glass = txtData_to_array('offline_data/offline_glass.txt')
scores_offline_ionosphere = txtData_to_array('offline_data/offline_ionosphere.txt')
scores_offline_pima = txtData_to_array('offline_data/offline_pima.txt')
scores_offline_mnist = txtData_to_array('offline_data/offline_mnist.txt')

scores_mondrian_wine = txtData_to_array('mondrian_data/mondrian_wine.txt')
scores_mondrian_lympho = txtData_to_array('mondrian_data/mondrian_lympho.txt')
scores_mondrian_glass = txtData_to_array('mondrian_data/mondrian_glass.txt')
scores_mondrian_ionosphere = txtData_to_array('mondrian_data/mondrian_ionosphere.txt')
scores_mondrian_pima = txtData_to_array('mondrian_data/mondrian_pima.txt')
scores_mondrian_mnist = txtData_to_array('mondrian_data/mondrian_mnist.txt')

scores_sk_wine = txtData_to_array('sk_data/sk_wine.txt')
scores_sk_lympho = txtData_to_array('sk_data/sk_lympho.txt')
scores_sk_glass = txtData_to_array('sk_data/sk_glass.txt')
scores_sk_ionosphere = txtData_to_array('sk_data/sk_ionosphere.txt')
scores_sk_pima = txtData_to_array('sk_data/sk_pima.txt')
scores_sk_mnist = txtData_to_array('sk_data/sk_mnist.txt')

sensory_data_scores = txtData_to_array('sensoryDataScores.txt')
```

```python

fig = plt.figure(figsize=(7, 5))
plt.title('Anomaly scores of gathered magnetometer data')
plt.xlabel('Data point')
plt.ylabel('Anomaly score')
sns.lineplot(data = sensory_data_scores)
plt.savefig('Magnetometer_line_scores.jpg')
```
